```python
from matplotlib import pyplot as plt
from matplotlib import style

#Plotting to our canvas
plt.plot([1,2,3],[4,5,1])

#Showing what we plotted
#plt.show()
```

```python
style.use('ggplot')
a=[3,4,5]
b=[8,4,3]
c=[6,7,4]
plt.plot(a,b,linewidth=10,label='mine')
plt.plot(b,c,linewidth=5,label='yours')
plt.title('testing')
plt.xlabel('height')
plt.ylabel('weight')
plt.legend()
plt.grid(True,color='k')
```

```python
plt.bar(a,b,label='mine')
plt.bar(b,c,label='yours')
plt.legend()
plt.title('testing')
plt.xlabel('height')
plt.ylabel('weight')
plt.grid(True,color='k')
```

```python
plt.scatter(a,b,label='mine')
plt.scatter(b,c,label='yours')
plt.legend()
plt.title('testing')
plt.xlabel('height')
plt.ylabel('weight')
plt.grid(True,color='k')
```
